```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Define the term-document matrix as a pandas DataFrame
data = {
    'As You Like It': [1, 114, 36, 20],
    'Twelfth Night': [0, 80, 58, 15],
    'Julius Caesar': [7, 62, 1, 2],
    'Henry V': [13, 89, 4, 3]
}
```

```python
# Terms (rows in the matrix)
terms = ['battle', 'good', 'fool', 'wit']
```

```python
# Create a DataFrame for the term-document matrix
term_document_matrix = pd.DataFrame(data, index=terms)
```

```python
# Display the term-document matrix
print("Term-Document Matrix:\n", term_document_matrix)
```

```python
# Calculate cosine similarity between the document vectors
cosine_sim = cosine_similarity(term_document_matrix.T)
cosine_sim_df = pd.DataFrame(cosine_sim, index=term_document_matrix.columns, columns=term_document_matrix.columns)
```
