```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
automobile_df = pd.read_csv("/content/sample_data/auto-mpg-processed.csv")
automobile_df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X = automobile_df[['horsepower']]
Y = automobile_df['mpg']
x_train,x_test, y_train,y_test = train_test_split(X,Y, test_size=0.2)
```

```python
x_train.sample(5)
```

The normalize parameter in the LinearRegression class of scikit-learn has been deprecated in version 0.24 and will be removed in future versions. Instead of using the normalize=True argument, you can normalize the features manually using StandardScaler from sklearn.preprocessing before fitting the model.

Here's how you can do it:

Import the necessary modules.
Normalize the features using StandardScaler.
Fit the linear regression model.
Here is an example:

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Assuming x_train and y_train are your training data

# Create a pipeline that standardizes the data then fits the model
pipeline = make_pipeline(StandardScaler(), LinearRegression())

# Fit the model
pipeline.fit(x_train, y_train)
```

```python
print('Training score:', pipeline.score(x_train, y_train))
```

```python
y_pred = pipeline.predict(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python

# Print the testing score
print('Testing score:', r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')
plt.xlabel('Horsepower')
plt.ylabel('Mpg')
```

```python
# Assuming 'automobile_df' is already defined and loaded
X = automobile_df[['age']]
Y = automobile_df['mpg']

# Split the data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Fit the linear model
linear_model = LinearRegression().fit(x_train, y_train)

# Print the training score
print('Training score:', linear_model.score(x_train, y_train))

# Predict and print the testing score
y_pred = linear_model.predict(x_test)
print('Testing score:', r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')
plt.xlabel('Age')
plt.ylabel('Mpg')
```

# Performing linear regression using multiple features

```python
X = automobile_df[['displacement', 'horsepower', 'weight']]
Y = automobile_df['mpg']
```

```python
# Split the data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Fit the linear model
linear_model = LinearRegression().fit(x_train, y_train)
```

```python
# Print the training score
print('Training score:', linear_model.score(x_train, y_train))
```

```python
predictors = x_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()
print(coef)
```

```python
# Predict and print the testing score
y_pred = linear_model.predict(x_test)
print('Testing score:', r2_score(y_test, y_pred))
```

```python
plt.figure(figsize = (20, 10))

plt.plot(y_pred, label='predicted')
plt.plot(y_test.values, label='Actual')

plt.ylabel('mpg')

plt.legend()
plt.show()
```
