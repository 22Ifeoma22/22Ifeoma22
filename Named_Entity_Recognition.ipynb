```python
# Ensure Pandas is installed or upgraded to the latest version
# Ensure Pandas is installed or upgraded to the latest version
import pandas as pd
# Removed the trailing slash from the file path
data = pd.read_csv("/content/sample_data/ner_datasetreference.csv", encoding="latin1")
data.head()
```

```python
data =data.fillna(method="ffill")
data.head()
```

```python
data.head(30)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data["Sentence #"] = LabelEncoder().fit_transform(data["Sentence #"])
```

```python
data.head(30)
```

```python
data.rename(columns={"Sentences #" : "sentences_id", "Word" : "words", "Tag" : "labels"}, inplace = True)
```

```python
data["labels"] = data["labels"].str.upper()
```

```python
# Correct the column name in the rename operation
data.rename(columns={"Sentence #" : "sentences_id", "Word" : "words", "Tag" : "labels"}, inplace = True)

# Now you can access the renamed columns
X = data[["sentences_id","words"]]
Y = data["labels"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
#building up train data and test data
train_data = pd.DataFrame({"sentences_id":x_train["sentences_id"], "words":x_train["words"], "labels":y_train})
test_data = pd.DataFrame({"sentences_id":x_test["sentences_id"], "words":x_test["words"], "labels":y_test})
```

```python
train_data.head()
```

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label = data["labels"].unique().tolist()
```

```python
label
```

```python
from simpletransformers.ner import NERModel, NERArgs

# ... (your other code) ...

args = NERArgs()
args.num_train_epochs = 1
args.learning_rate = 1e-4
args.overwrite_output_dir = True
args.train_batch_size = 32
args.eval_batch_size = 32
# Explicitly set use_cuda to False to force CPU usage
args.use_cuda = False # This line disables CUDA usage

```

```python
import torch
print(torch.cuda.is_available())  # Should print True if a GPU is available
```

```python
args.use_cuda = False  # Ensure it uses CPU if CUDA is not available
```

```python
model = NERModel('bert', 'bert-base-cased', labels=label, args=args)
```

```python
import pandas as pd

# Assuming your train_data is a pandas DataFrame
# Add a 'sentence_id' column if it doesn't exist
if 'sentence_id' not in train_data.columns:
    # If you have a column that groups sentences, rename it to 'sentence_id'
    # For example, if you have a column named 'group_id':
    # train_data = train_data.rename(columns={'group_id': 'sentence_id'})
    # Otherwise, create a new 'sentence_id' column based on your data structure.
    # Below is a basic example assuming each row represents a separate sentence:
    train_data['sentence_id'] = train_data.index


# Now you can train the model
model.train_model(train_data, eval_data=test_data, acc=accuracy_score)
```

```python
import pandas as pd

# Assuming your test_data is a pandas DataFrame
# Add a 'sentence_id' column if it doesn't exist
if 'sentence_id' not in test_data.columns:
    # If you have a column that groups sentences, rename it to 'sentence_id'
    # For example, if you have a column named 'sentences_id' (as seen in global variables)
    # test_data = test_data.rename(columns={'sentences_id': 'sentence_id'})
    # Otherwise, create a new 'sentence_id' column based on your data structure.
    # Below is a basic example assuming each row represents a separate sentence:
    test_data['sentence_id'] = test_data.index

# Now you can evaluate the model
result, model_outputs, predictions = model.eval_model(test_data)
```
