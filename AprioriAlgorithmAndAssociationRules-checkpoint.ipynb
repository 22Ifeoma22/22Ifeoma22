```python
items_per_transaction = items.groupby('Transaction').agg(','.join).reset_index()
    
items_per_transaction.head(5)
```

```python
items_per_transaction = items_per_transaction.drop(
['Date', 'Time', 'year_month', 'day', 'weekend_vs_weekday'], axis=1)

items_per_transaction.head(5)
```

```python
items_per_transaction.head(5)
```

```python
items_per_transaction['Item']
```

```python
items_list = [item.split(',') for item in items_per_transaction['Item']]

items_list
   
```

```python
items_list_df = pd.DataFrame({'Items': items_list})

items_list_df.head(10)
```

```python
from mlxtend.preprocessing import TransactionEncoder

transencoder = TransactionEncoder()
transencoder_array = transencoder.fit(items_list).transform(items_list)
```

```python
items_df = pd.DataFrame(transencoder_array, columns=transencoder.columns_)
    
items_df.sample(10)
```

```python
items_df.shape
```

```python
coffee_df = items_df[items_df['Coffee'] == True]
coffee_df
```

```python
coffee_df['Coffee']
```

```python
bread_df = items_df[items_df['Bread'] == True]
```

```python
bread_df
```

```python
jam_coffee_bread_df = items_df[(items_df['Jam'] == True) & \
                    (items_df['Coffee'] == True) &\
                     (items_df['Bread'] == True)]
jam_coffee_bread_df                               
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
items_frequent = apriori(items_df,
                          min_support = 0.09,  # 10% in the items
                          use_colnames = True)
    
items_frequent
```

```python
items_frequent = apriori(items_df,
                          min_support = 0.1,  # 10% in the items
                          use_colnames = True)
    
items_frequent
```

```python
items_frequent = apriori(items_df,
                          min_support = 0.05,  # 10% in the items
                          use_colnames = True)
    
items_frequent
```

```python
items_frequent = apriori(items_df,
                          min_support = 0.01,  # 10% in the items
                          use_colnames = True)
    
items_frequent
```

# Association Rules Mining

```python
association = association_rules(items_frequent,
                                  metric = "confidence",
                                  min_threshold = 0.5) i# items is 50%
    
association
```

```python
association = association_rules(items_frequent,
                                  metric = "confidence",
                                  min_threshold = 0.3) # items is 30%
    
association
```

```python
association = association_rules(items_frequent,
                                  metric = "lift",
                                  min_threshold =  1) 

    
association
```

```python
association[association['confidence'] >= 0.5].sort_values('confidence', 
                                                        ascending = False).head(10)
```
